```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df= pd.read_excel(r"C:/Users/User 1/OneDrive/Desktop/Dots Potential Customer Survey Data.xlsx")
df.head()
```

```python
categorical_vars = df.select_dtypes(include=['object']).columns.tolist()
```

```python
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()
for col in categorical_vars:
    df[col] = label_encoder.fit_transform(df[col])
```

```python
outliers = pd.DataFrame()
```

```python
for country in df['Country'].unique():
    country_df = df[df['Country'] == country]
    Q1 = country_df.quantile(0.25)
    Q3 = country_df.quantile(0.75)
    IQR = Q3 - Q1
    country_outliers = country_df[(country_df < (Q1 - 1.5 * IQR)) | (country_df > (Q3 + 1.5 * IQR))].dropna()
    outliers = pd.concat([outliers, country_outliers])
```

```python
outliers.reset_index(drop=True, inplace=True)
```

```python
numerical_vars = df.select_dtypes(include=['int64', 'float64']).columns.tolist()
for col in numerical_vars:
    plt.figure(figsize=(8, 6))
    sns.histplot(df[col], bins=20, kde=True)
    plt.title(f'Histogram of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.show()
```

```python
summary_stats = df.describe()
```

```python
correlation_matrix = df.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Correlation Matrix')
plt.show()
```
