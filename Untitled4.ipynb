```python
correlation_matrix = df.corr()
correlation_with_target = correlation_matrix['Wear Specs']
```

```python
threshold = 0.2  # Adjust this threshold as per your dataset
relevant_columns = correlation_with_target[abs(correlation_with_target) >= threshold].index
df_relevant = df[relevant_columns]
```

```python
# Identify cutoffs for each remaining column
for column in df_relevant.columns:
    if df_relevant[column].nunique() > 10:  # Skip if too many unique values
        continue
    value_counts = df_relevant[column].value_counts(normalize=True)
    cutoff_value = value_counts[value_counts.cumsum() >= 0.7].index[0]
    print(f"Cutoff for {column}: {cutoff_value}")
```

```python
binary_columns = [col for col in df_relevant.columns if df_relevant[col].nunique() == 2]

for column in binary_columns:
    spec_grouped = df_relevant.groupby(column)['Age'].mean()
    difference = spec_grouped.diff().abs().max()
    print(f"Difference in chances of getting specs between groups in {column}: {difference}")
```
